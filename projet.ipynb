# Partie 1.b

### 5x10

```python
list_discount_factor = np.linspace(0.5,1.0,1000)
nb_test = 100

list_time_mean = list()
list_time_std = list()


for discount_factor in tqdm(list_discount_factor) :
    
    list_tmp = list()
    
    for t in range(nb_test) :
    
        time_training = value_iteration_test_mdp(
            discount_factor = discount_factor,
            nb_lignes = 5,
            nb_colonnes = 10,
            alea = 0 # deterministe
        )
        
        list_tmp.append(time_training)
        
    list_tmp = np.array(list_tmp)
    list_time_mean.append( list_tmp.mean() )
    list_time_std.append( list_tmp.std() ) 
```

```python
plt.figure(figsize=(10,7))
list_mean1_3 = np.array(list_time_mean) + np.array(list_time_std)
list_mean2_3 = np.array(list_time_mean) - np.array(list_time_std)
plt.fill_between(list_discount_factor,list_mean1_3,list_mean2_3, color = 'salmon', label = 'std time (s)')
plt.plot(list_discount_factor,list_time_mean, c= 'r',label = 'mean time (s)')
plt.legend()
plt.xlabel('discount factor')
plt.ylabel('time (s)')
plt.title('Temps d\' apprentissage d\' MDP 5x10 en fonction du discount factor')
plt.show()
```

# Partie 1.c

```python
nb_tests = 10
tailles = [(8,8),(10,10),(11,11),(12,12),(13,13),(14,14),(15,15),(16,16),(17,17),(18,18),(19,19),(20,20)]
discount_factors = [0.9,0.7,0.5]

results = dict()

for taille in tqdm(tailles) :
    
    for discount_factor in discount_factors :
        
        list_time = list()
        list_iteration = list()
        
        for _ in range(nb_tests) :
            
            time_training,iteration_training = value_iteration_test_mdp_partie_c(
                discount_factor = discount_factor,
                nb_lignes = taille[0],
                nb_colonnes = taille[1],
                alea = 1 # stochastique
            )
            list_time.append(time_training)
            list_iteration.append(iteration_training)
           
        list_time = np.array(list_time)
        list_iteration = np.array(list_iteration)
        
        mean_time = list_time.mean()
        std_time = list_time.std()
        mean_iteration = list_iteration.mean()
        std_iteration = list_iteration.std()
        
        
        print(f"taille : {taille} - discount factor : {discount_factor} - mean time : {mean_time}s +/- {std_time} - iteration mean : {mean_iteration} +/- {std_iteration}")
        
        # sauvegarde des résultats
        
        results[(taille,discount_factor)] = (mean_time,std_time,mean_iteration,std_iteration)
```
